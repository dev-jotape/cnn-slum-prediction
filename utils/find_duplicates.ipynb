```python
import os
import geopandas as gpd
```

```python
grid = gpd.read_file('../QGIS/aglomerados_subnormais/shapefile_sampled_data/shapefile_sampled_data.shp')
grid
```

```python
def encontrar_arquivos_duplicados(diretorio):
    verificados = []
    duplicados = []

    for _, _, arquivos in os.walk(diretorio):
        print(len(arquivos))
        for nome_arquivo in arquivos:
            fid = int(nome_arquivo.split('_')[0])

            if fid in verificados:
                duplicados.append(fid)
            else:
                verificados.append(fid)

            # find in grid
            if fid in grid['fid'].values:
                verificados.append(fid)
            else:
                print(f'fid {fid} not found')
                
    # print(verificados)
    return duplicados

# Exemplo de uso
diretorio = './GEE_SENT2_RGB_2020_05'
arquivos_duplicados = encontrar_arquivos_duplicados(diretorio)

if arquivos_duplicados:
    print("Arquivos duplicados encontrados:")
    for original, duplicado in arquivos_duplicados:
        print(f"Original: {original} | Duplicado: {duplicado}")
else:
    print("Nenhum arquivo duplicado encontrado.")
```
