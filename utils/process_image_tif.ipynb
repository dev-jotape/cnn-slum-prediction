```python
import rasterio
import os
```

```python
folder = '../dataset/GEE_SENT2_RGB_2020_05'

for _, _, files in os.walk(folder):
        print(len(files))
        for file_path in files:
            print(file_path)
            with rasterio.open(folder + '/' + file_path) as src:
                num_bands = src.count
                print(f"O arquivo TIFF tem {num_bands} bandas.")
```
