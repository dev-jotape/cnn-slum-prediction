```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Matriz de F1 Scores (exemplo)
f1_scores = np.array([
    # SP    RJ    BH    BR   SSA    PA
    [0.93, 0.87, 0.90, 0.81, 0.78, 0.75],  # SP
    [0.86, 0.91, 0.87, 0.76, 0.80, 0.76],  # RJ
    [0.86, 0.85, 0.93, 0.80, 0.74, 0.76],  # BH
    [0.78, 0.78, 0.77, 0.90, 0.72, 0.69],  # BR
    [0.82, 0.83, 0.83, 0.83, 0.85, 0.74],  # SSA
    [0.77, 0.83, 0.79, 0.71, 0.76, 0.81],  # PA
])

# Nome das cidades
cities = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Brasília', 'Salvador', 'Porto Alegre']
```

```python
# Criar heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(f1_scores, annot=True, cmap="YlGnBu", xticklabels=cities, yticklabels=cities)

# plt.title('F1 Score Heatmap')
# plt.xlabel('Cidade Avaliada')
# plt.ylabel('Cidade Treinada')
plt.show()
```
