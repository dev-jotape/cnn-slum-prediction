```python
import rioxarray
import pandas as pd
import geopandas as gpd
```

```python
rds = rioxarray.open_rasterio("../../QGIS/aglomerados_subnormais/ntl_sp.tif", cache=False)
rds.name = "data"
```

```python
xmin, ymin, xmax, ymax = [-74.00, -34.00, -33.00, 6.00]
df = rds.squeeze().sel(x=slice(xmin, xmax)).to_dataframe().reset_index()
df
```

```python
nighttime_lights = df[['y','x','data']].rename({'y':'lat', 'x':'long', 'data': 'radiance'}, axis=1)
nighttime_lights
```

```python
shapefile_br = gpd.read_file('../../QGIS/aglomerados_subnormais/sp_grid_area/grid.shp')
shapefile_br
```

```python
gdf = gpd.GeoDataFrame(
      nighttime_lights, geometry=gpd.points_from_xy(nighttime_lights.long, nighttime_lights.lat))
gdf
```

```python
# points_in_polygons_gdf = gpd.sjoin(gdf, shapefile_br, predicate='within', how='inner')
# points_in_polygons_gdf
# points_gdf = gdf.to_crs(shapefile_br.crs)
joined_gdf = gpd.sjoin(gdf, shapefile_br, how="inner", op="intersects")

# Calcular as estatísticas por célula do grid
stats_df = joined_gdf.groupby('index_right').agg(
    count=('radiance', 'count'),
    sum=('radiance', 'sum'),
    mean=('radiance', 'mean')
).reset_index()

# Unir as estatísticas calculadas ao GeoDataFrame dos grids
grids_gdf = shapefile_br.merge(stats_df, left_index=True, right_on='index_right', how='left')

# Preencher valores NaN com 0 (caso não haja pontos em algumas células)
grids_gdf[['count', 'sum', 'mean']] = grids_gdf[['count', 'sum', 'mean']].fillna(0)
grids_gdf
```

```python
grids_gdf.to_file('../dataset/nightlights_sp/nightlights_sp.shp', index=False)
```
