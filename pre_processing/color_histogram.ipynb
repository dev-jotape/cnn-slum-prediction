```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

```

```python
# Função para calcular e plotar o histograma de cores
def plot_rgb_histogram(image_path):
    # Abrindo a imagem usando Pillow
    img = Image.open(image_path)
    print(img.mode)
    img = img.convert('RGB')

    # Certifique-se de que a imagem está em formato RGB de 16 bits
    # if img.mode != 'I;16':
    #     raise ValueError("A imagem precisa ser em formato RGB com 16 bits por canal ('I;16').")
   
    # Converter a imagem para array NumPy
    img_np = np.array(img)
   
    # Separar os canais R, G e B
    r, g, b = [np.histogram(img_np[..., i], bins=256, range=(0, 256))[0] for i in range(3)]

    # Definir o número de bins (0 até 2^16-1)
    # bins = np.arange(0, 2**16+1)

    # Criar o gráfico de histogramas
    plt.figure(figsize=(12, 6))
   
    # Histograma para o canal vermelho
    plt.subplot(3, 1, 1)
    plt.title('Histograma de Cores')
    plt.bar(range(256), r, color='red', alpha=0.6)
    plt.xlim([0, 255])
    plt.ylabel('Frequência')

    plt.subplot(3, 1, 2)
    plt.bar(range(256), g, color='green', alpha=0.6)
    plt.xlim([0, 255])
    plt.ylabel('Frequência')

    plt.subplot(3, 1, 3)
    plt.bar(range(256), b, color='blue', alpha=0.6)
    plt.xlim([0, 255])
    plt.ylabel('Frequência')

    plt.xlabel('Valor da Cor')
    plt.tight_layout()
    plt.show()
```

```python

# Caminho da imagem
# image_path = 'sua_imagem_16bits.png'
image_paths_slum = "../dataset/GMAPS_RGB_SP_2024/4450_1.png"

# Plotar os histogramas
plot_rgb_histogram(image_paths_slum)
```
