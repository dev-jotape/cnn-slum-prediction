```python
import geopandas as gpd
from sklearn.model_selection import train_test_split
```

```python
df = gpd.read_file('../../QGIS/aglomerados_subnormais/final_shapefiles/sp/shapefile_sampled_data.shp')
df
```

```python
_, stratified_sample = train_test_split(df, test_size=3000, stratify=df[['class']])
stratified_sample
```

```python
ids = stratified_sample['fid'].astype(int).tolist()
print(len(ids))
```

```python
# filter images based on ids from dataset/old/GMAPS_RGB_SP_2024 to dataset/GMAPS_RGB_SP_2024

import os
import shutil

src = '../dataset/old/GMAPS_RGB_SP_2024'
dst = '../dataset/GMAPS_RGB_SP_2024'

for idx, id in enumerate(ids):
    # print(idx)
    achou = False
    for filename in os.listdir(src):
        if filename.startswith(str(id)):
            shutil.copy(os.path.join(src, filename), dst)
            # print(f'Copied {filename} to {dst}')
            achou = True
            break
        else:
            # print(f'Image for id {id} not found')
            continue
    if not achou:
        print(f'Image for id {id} not found')
```
